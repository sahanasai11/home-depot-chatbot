```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from chromedriver_py import binary_path # this will get you the path variable
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
urls = pd.read_csv('appliance_urls')
all_product_links = set()

# appliances will not show up bc of different HTML structure
errors = set()
MAX_PRODUCTS_PER_PAGE = 4


# Selenium configs
options = Options()
options.add_argument('--headless=new')
options.add_argument('--window-size=1920,1200')
service_object = Service(binary_path)
driver = webdriver.Chrome(options=options, service=service_object)
```

```python
# finds links that will lead to product detail page
xpath_products_page = "//div[contains(@data-testid, 'product-pod')]/a"

length = urls.shape[0] 

count = 0
for i, row in urls.iterrows():
    if count % 100 == 0:
        print('Progress: {}%'.format(count/length))
    
    count += 1
        
    url = row['url']
    driver.get(url)
    
    # wait for page to load
    product_link = driver.find_elements(By.XPATH, xpath_products_page)
    
    if product_link == []:
        errors.add(url)
        print(url)
        
    # ensure that if there are less products than MAX_PRODUCTS_PER_PAGE, no out of bounds error
    for j in range(min(MAX_PRODUCTS_PER_PAGE, len(product_link))):
        try:
            all_product_links.add(product_link[j].get_attribute('href'))
        except:
            errors.add(url)
            
driver.quit()      
        
```

```python
product_urls = pd.read_csv('product_urls')

appliances_product_urls = pd.DataFrame(all_product_links, columns=['url'])

full_product_urls = pd.concat([product_urls, appliances_product_urls])
```
